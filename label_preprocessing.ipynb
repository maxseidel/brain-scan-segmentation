```python
import os

# Get results csv filenames
results = os.listdir("segmentation_labels")
results = [r for r in results if r.startswith("Results")]
results
```

```python
invalid_ids = []

for r_path in results:
    results_df = pd.read_csv("segmentation_labels/" + r_path)
    ids = list(results_df[(results_df["Majority Label"].apply(len) == 2) & (results_df["Correct Label"].isnull())]["Origin"])
    ids = [id.split(".")[0] for id in ids]
    invalid_ids.extend(ids)
invalid_ids
    
```

```python
import csv

with open("segmentation_labels/invalid_ids.csv", 'w', newline='') as file:
   wr = csv.writer(file)
   for id in invalid_ids:
      print(id)
      wr.writerow([id])
```
